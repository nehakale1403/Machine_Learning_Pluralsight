```python
import matplotlib.pyplot as plt

from skimage import data, io, color, segmentation
from skimage.future import graph
from skimage import io
```

```python
nature = io.imread('images/nature.jpeg')

plt.figure(figsize=(8,8))
plt.imshow(nature, cmap='gray')
```

```python
labels_1 = segmentation.slic(nature, compactness=35, n_segments=500)

labels_1.shape
```

```python
labels_1
```

```python
segmented_overlay = color.label2rgb(labels_1, nature, kind='overlay')

plt.figure(figsize=(8,8))
plt.imshow(segmented_overlay)
```

```python
segmented_overlay = color.label2rgb(labels_1, nature, kind='avg')

plt.figure(figsize=(8,8))
plt.imshow(segmented_overlay)
```

```python
g = graph.rag_mean_color(nature, labels_1)
```

```python
labels_2 = graph.cut_threshold(labels_1, g, thresh=35)
```

```python
segmented_rag = color.label2rgb(labels_2, nature, kind='avg')

plt.figure(figsize=(8,8))
plt.imshow(segmented_rag)
```
