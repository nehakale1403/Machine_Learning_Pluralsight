```python
import matplotlib.pyplot as plt

from skimage.morphology import convex_hull_image
from skimage import data, img_as_float
from skimage.util import invert
```

```python
horse_image = data.horse()

horse_inverted = invert(data.horse())# white portion becomes black and black becomes white
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,12), sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(horse_image, cmap='gray')
ax[0].set_title('Original', fontsize=20)

ax[1].imshow(horse_inverted, cmap='gray')
ax[1].set_title('Inverted', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
convexhull_horse = convex_hull_image(horse_inverted)
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,12), sharex=True, sharey=True)

ax = axes.ravel()

ax[0].imshow(horse_inverted, cmap='gray')
ax[0].set_title('Inverted', fontsize=20)

ax[1].imshow(convexhull_horse, cmap='gray')
ax[1].set_title('Inverted', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
convexhull_diff = img_as_float(convexhull_horse.copy())

convexhull_diff[horse_inverted] = 2
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.imshow(convexhull_diff, cmap='gray', interpolation='nearest')

ax.set_title('Difference between Inverted and Convex Hull')

plt.show()
```
