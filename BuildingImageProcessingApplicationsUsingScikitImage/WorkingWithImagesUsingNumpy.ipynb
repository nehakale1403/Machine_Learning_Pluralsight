```python
tree_copy = tree.copy()

mask = tree_copy < 0.5
```

```python
mask
```

```python
tree_copy[mask] = 255 #where the intensity of pixel were less that 0.5, we r setting it to 255

plt.figure(figsize=(8,6))

plt.imshow(tree_copy, cmap='gray')
```

```python
tree_copy = tree.copy()

mask=tree_copy<0.5

tree_copy[mask] = 0

plt.figure(figsize=(8,6))
plt.imshow(tree_copy, cmap='gray')
```

```python
parrot = skimage.io.imread('images/parrot.jpeg')
```

```python
plt.figure(figsize=(6,6))

plt.imshow(parrot, interpolation='nearest')
```

```python
parrot.shape
```

```python
parrot
```

```python
blue_mask = parrot[:,:,2] < 15 #the mask finds those pixels where blue<15

blue_mask
```

```python
parrot_copy = parrot.copy()

parrot_copy[blue_mask] = [0,0,255]

plt.figure(figsize=(8,8))

plt.imshow(parrot_copy)
```

```python
red_mask = parrot[:,:,0]<15

red_mask
```

```python
parrot_copy = parrot.copy()

parrot_copy[red_mask] = [255,0,0]

plt.figure(figsize=(8,8))

plt.imshow(parrot_copy)
```

```python
parrot_zoom = parrot[100:300, 200:400]

plt.imshow(parrot_zoom)
```
