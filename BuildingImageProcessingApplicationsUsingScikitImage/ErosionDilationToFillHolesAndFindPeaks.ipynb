```python
import matplotlib.pyplot as plt
import numpy as np

from skimage import io, data

from skimage.exposure import rescale_intensity
from skimage.morphology import reconstruction
```

```python
moon = io.imread('images/moon1.jpg')

plt.figure(figsize=(8,8))
plt.imshow(moon, cmap='gray')
```

```python
moon_rescaled = rescale_intensity(moon, in_range=(30, 200))

plt.figure(figsize=(8,8))
plt.imshow(moon_rescaled, cmap='gray')
```

```python
erosion_seed = np.copy(moon_rescaled)

erosion_seed[1:-1, 1:-1] = moon_rescaled.max()

mask=moon_rescaled
```

```python
#erosion fills the holes
filled = reconstruction(erosion_seed, mask, method='erosion')
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,12), sharex=True, sharey=True)

ax=ax.ravel()

ax[0].imshow(moon_rescaled, cmap='gray')
ax[0].axis('off')
ax[0].set_title('Original image', fontsize=20)

ax[1].imshow(filled, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Erosion reconstruction', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
dilation_seed = np.copy(moon_rescaled)

dilation_seed[1:-1, 1:-1] = moon_rescaled.min()

mask=moon_rescaled
```

```python
highlighted = reconstruction(dilation_seed, mask, method='dilation')
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,12), sharex=True, sharey=True)

ax=ax.ravel()

ax[0].imshow(moon_rescaled, cmap='gray')
ax[0].axis('off')
ax[0].set_title('Original image', fontsize=20)

ax[1].imshow(highlighted, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Dilated reconstruction', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
holes= filled-moon_rescaled

peaks= highlighted - moon_rescaled
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,12), sharex=True, sharey=True)

ax=ax.ravel()

ax[0].imshow(holes, cmap='gray')
ax[0].axis('off')
ax[0].set_title('Holes', fontsize=20)

ax[1].imshow(peaks, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Peaks', fontsize=20)

fig.tight_layout()
plt.show()
```
