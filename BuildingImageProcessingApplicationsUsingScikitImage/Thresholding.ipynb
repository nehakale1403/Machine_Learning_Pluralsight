```python
import matplotlib.pyplot as plt

from skimage import data, io, color
from skimage.filters import threshold_otsu, threshold_local
```

```python
cathedral = color.rgb2gray(io.imread('images/church.jpg'))

plt.figure(figsize=(8,8))
plt.imshow(cathedral, cmap='gray')
```

```python
# scaling
cathedral  = cathedral*255 
```

```python
thresh = threshold_local(cathedral, block_size=25)

print('Threshold  value of this image', thresh)
```

```python
cathedral_binary = cathedral>thresh
```

```python
print('binary values of the image ', cathedral_binary)
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,12), sharex=True, sharey=True)

ax=ax.ravel()

ax[0].imshow(cathedral, cmap='gray')
ax[0].axis('off')
ax[0].set_title('Original image', fontsize=20)

ax[1].imshow(cathedral_binary, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Thresholded', fontsize=20)

fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.hist(cathedral.ravel())

ax.set_title('Histogram')
ax.set_xlabel('Pixel intensity')
ax.set_ylabel('Pixel count')

ax.axvline(thresh, color='y')
```
