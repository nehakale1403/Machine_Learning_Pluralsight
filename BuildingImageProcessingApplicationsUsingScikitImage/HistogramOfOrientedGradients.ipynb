```python
import matplotlib.pyplot as plt

import skimage
from skimage.feature import hog
from skimage import data, color, io, exposure
```

```python
monalisa = skimage.io.imread('images/monalisa.jpg')

plt.figure(figsize=(10,8))
plt.imshow(monalisa, cmap='gray')
```

```python
fd, hog_image = hog(monalisa, pixels_per_cell=(8,8), block_norm = 'L2-Hys', visualize=True, multichannel=True)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,8), sharex=True, sharey=True)

ax=axes.ravel()

ax[0].imshow(monalisa, cmap='gray')
ax[0].set_title('Original')

hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0,10))

ax[1].imshow(hog_image_rescaled, cmap='gray')
ax[1].set_title('HOG image')

plt.tight_layout()
plt.show()
```
