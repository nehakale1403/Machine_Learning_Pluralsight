```python
import matplotlib.pyplot as plt

import skimage
from skimage import color
from skimage.feature import daisy
from skimage import data,io
```

```python
car = skimage.io.imread('images/car.jpeg')

car = color.rgb2gray(car)
plt.figure(figsize=(10,8))
plt.imshow(car, cmap='gray')
```

```python
descs_array, descs_image = daisy(car, step=200, radius = 20, visualize = True)
```

```python
descs_num = descs_array.shape[0] * descs_array.shape[1]

descs_num
```

```python
plt.figure(figsize=(12,10))

plt.title('DAISY Descriptpors', fontsize=18)
plt.imshow(descs_image)
```
