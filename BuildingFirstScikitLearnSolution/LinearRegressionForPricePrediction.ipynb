```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
housing_data = pd.read_csv('datasets/housing.csv')
housing_data.sample(5)
```

```python
housing_data = housing_data.dropna()
```

```python
housing_data.shape
```

```python
housing_data.loc[housing_data['median_house_value']==500001].count()
```

```python
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value']==500001].index)
```

```python
housing_data.shape
```

```python
housing_data.head()
```

```python
housing_data['ocean_proximity'].unique()
```

```python
#one-hot encoding for converting the text data to numeric data in the ocean_proximity column
housing_data = pd.get_dummies(housing_data, columns=['ocean_proximity'])
```

```python
housing_data.shape
```

```python
housing_data.sample(5)
```

```python
# x values are features that we use to train our ML model
# y values are our predictions 
X = housing_data.drop('median_house_value', axis = 1)
Y = housing_data['median_house_value']
```

```python
X.columns
```

```python
# we have to split the data into training data and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2)
#here 0.2 parts of the original dataset are reserved for the testing purpose
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
#normalization scales all numeric features to be between  0 and 1
#fit function fits Linear model on our training data
linear_model = LinearRegression(normalize = True).fit(x_train, y_train)
```

```python
print('Training score : ', linear_model.score(x_train, y_train))
# our model captures 61% of variations of our training data 
```

```python
predictors = x_train.columns
predictors
```

```python
# coefficient is the measure of how much a particular feature contributed towards the predictions
coef = pd.Series(linear_model.coef_, predictors).sort_values()

print(coef)
```

```python
y_pred = linear_model.predict(x_test)
```

```python
df_pred_actual = pd.DataFrame({'predicted': y_pred, 'actual' :y_test })

df_pred_actual.head(10)
```

```python
from sklearn.metrics import r2_score

print("Testing_score: ", r2_score(y_test, y_pred))
```

```python
fig, ax  = plt.subplots(figsize=(12,8))

plt.scatter(y_test, y_pred)
plt.show()
```

```python
df_pred_actual_sample = df_pred_actual.sample(100)

df_pred_actual_sample = df_pred_actual_sample.reset_index()
```

```python
df_pred_actual_sample.head()
```

```python
plt.figure(figsize = (20,10))

plt.plot(df_pred_actual_sample['predicted'], label = 'Predicted')
plt.plot(df_pred_actual_sample['actual'], label = 'Actual')

plt.ylabel('median_house_value')

plt.legend()
plt.show()
```
