```python
import pandas as pd
```

```python
data = pd.read_csv('datasets/exams.csv')
```

```python
data.head()
```

```python
math_average = data['math score'].mean()
reading_average = data['reading score'].mean()
writing_average = data['writing score'].mean()

print('Math avg: ', math_average)
print('Reding avg: ', reading_average)
print('Writing avg:',writing_average) 
```

```python
# this will standardize the data

from sklearn import preprocessing

data[['math score']] = preprocessing.scale(data[['math score']])
data[['reading score']] = preprocessing.scale(data[['reading score']])
data[['writing score']] = preprocessing.scale(data[['writing score']])
```

```python
data
```

```python
math_average = data['math score'].mean()
reading_average = data['reading score'].mean()
writing_average = data['writing score'].mean()

print('Math avg: ', math_average)
print('Reding avg: ', reading_average)
print('Writing avg:',writing_average) 
```
